# Facility

```python
import openai
import time

generated_sentences = set()

def generate_unique_sentences(messages):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    
    words = response.choices[0]['message']['content'].strip().split(', ')
    return words

base_messages = [
    {"role": "system", "content": "You are an AI model that generates unique sentences about movie theatre facilities. Avoid repetitions of previously generated sentences."},
]

for i in range(50):
    # Provide context from the last 100 generated sentences
    context = list(generated_sentences)[-10:]
    context_str = ', '.join(context)

    messages = base_messages + [
        {"role": "user", "content": f"Generate 20 new sentences. Context: {context_str}"}
    ]
    new_words = generate_unique_sentences(messages)
    
    for word in new_words:
        if len(generated_sentences) < 1000 and word not in generated_sentences:
            generated_sentences.add(word)

    # Pause between requests to respect rate limits
    time.sleep(21)  # Adjust the sleep time as needed based on your rate limits

with open("/content/drive/My Drive/NLP Project/sentence_facility_generation.txt", "w") as f:
    for word in generated_sentences:
        f.write(f"{word}\n")
```
