```python
import openai
import time

generated_sentences = []

def Aspect_Determination(messages):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    
    words = response.choices[0]['message']['content'].strip().split(', ')
    return words

evaluaion_list = []

base_messages = [
    {"role": "system", "content": "You are an AI model that determine if the paragraph about movie reviews has the following 4 aspects:\
    movie plots; actor & actress performance; movie theatre facility; movie special effects or stage design. Paragraphs seem like talking about the content of the movie should be 'movie plots'.\
Only when none of the 4 aspects can be determined, then output the singly 'others', others should not be combined with other aspects in the output."},
]

for text in evaluation_sentence:

    messages = base_messages + [
        {"role": "user", "content": f"Determine aspects of paragraph, the output format should only be aspect names, for example, 'movie plots, movie theatre facility': Context: {text}"}
    ]
    generated_aspect = Aspect_Determination(messages)

    evaluaion_list.append(generated_aspect)
    # Pause between requests to respect rate limits
    time.sleep(21)  # Adjust the sleep time as needed based on your rate limits

with open("/content/drive/My Drive/NLP Project/evaluation.txt", "w") as evaluation:
    for line in evaluaion_list:
        evaluation.write(f"{line}\n")
```
